```python
import pandas as pd 
import numpy as np 

from pandas import Series, DataFrame
```

```python
np.random.seed(25)
# função utilizada para gerar a semente do gerador de números aleatórios em npy.

# criando os indices (linhas) e as colunas com os dados.

indice = [ 'linha 1', 'linha 2', 'linha 3', 'linha 4','linha 5', 'linha 6']
colunas = ['coluna 1', 'coluna 2', 'coluna 3', 'coluna 4', 'coluna 5', 'coluna 6']

df = DataFrame(np.random.rand(36).reshape((6,6)),
               
## definindo o objeto 'df' como DataFrame 
# np.random.rand(n) cria uma matriz de algoritmo aleatório 
# 'n' é a quantidade de valores 
# reshape 'reorganiza' os dados -> ((6,6)) significa: 6 linhas, 6 colunas

index=indice,
columns=colunas)

# invocando o objeto 'df'

df 
```

```python
# execultando comparações entre o nosso DataFrame e um randon value. 
# quais valores nesse 'df' são menores do que .2? 

df < .2 

# me retorna um dataframe booleano (true ou false)
# o python execultou para cada número uma comparação 
```

```python
# utilizando comparações para criar filtros 

indice = ['linha 1', 'linha 2', 'linha 3', 'linha 4',
          'linha 5', 'linha 6', 'linha 7', 'linha 8']

series_obj = Series(np.arange(8), index=indice)

# np.arange(8) = retorna valores uniformemente espaçados dentro de um determinado intervalo.
```

```python
filtro = series_obj > 6 
series_obj[filtro]
```

```python
# atualizando valores 
series_obj['linha 1', 'linha 5', 'linha 7'] = 8 
series_obj 

# a esquerda: ídices, a direita: valores atualizados. 
```
