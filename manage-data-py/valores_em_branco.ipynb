```python
# valores em branco com Python:
# por padrão, valores em brancos são vistos como NaN 
```

```python
import numpy as np 
import pandas as pd 

from pandas import Series, DataFrame 
```

```python
# o numpy possui um objeto que define valores em branco (NaN)

white_file = np.nan 

serie = Series(['linha 1', 'linha 2', white_file, 'linha 4', 'linha 5', white_file, 'linha 6', white_file, 'linha 8'])

serie 
```

```python
serie.isnull()

# compara os números e aponta 'true' para valores vazios 
```

```python
np.random.seed(25)
# gerando 36 valores aleatórios
df = DataFrame(np.random.rand(36).reshape(6,6))

# invocando o objeto 'df'
# trocando seus valores 
#df[3:5, 0] = da linha 3 até a linha 5, na coluna 0 irá ser 'shite_file' (NaN)

df[3:5, 0] = white_file 
df[1:4, 5] = white_file 
```

```python
# .loc = acessa um grupo de linhas e colunas por rótulo(s) ou uma matriz booleana.

df.loc[3:5, 0] = white_file 
df.loc[1:4, 5] = white_file 

df 
```

```python
# preencher os valores NaN com o método .fillan

df_full = df.fillna(0)
df_full

# NaN foi preenchido com '0'
```

```python
dicio = {0: 0.1, 5: 1.25}
# dicionarios em python: coleção com elementos chave-valor
# semelhante ao objeto em JS
df_full = df.fillna(dicio)

df_full 
```

```python
# contando e removendo elementos NaN 
```

```python
white_file = np.nan 
np.random.seed(25)
df = DataFrame(np.random.rand(36).reshape(6,6))
df.loc[3:5, 0] = white_file 
df.loc[1:4, 5] = white_file 

df
```

```python
# contando os valores em branco com:
# .isnull() = identificar os valores em branco
# .sum() = fazer a soma de todos os valores em branco e retornar um numero

df.isnull().sum()

# coluna 0 = 3 valores NaN
# coluna 5 = 4 valores NaN
```

```python
# removendo valores em branco com:
# .dropna()

df.dropna()

# ele me retorna todas as colunas apagadas
# dessa forma, resultando um erro
# para utilizar um filtro no drop: 

# dropna(axis=1)
```

```python
df.dropna(axis=1)

# um novo DataFrame que não apresenta colunas com valores em branco
```
